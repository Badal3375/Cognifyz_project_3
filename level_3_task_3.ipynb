```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import  seaborn as sns
```

```
df=pd.read_csv('/content/Dataset .csv')
```

```
df.head(1)
```

```
plt.figure(figsize=(10, 5))
sns.histplot(df['Aggregate rating'].dropna(), bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

*** Using Bar Plot (Count of each rating value)***

```
plt.figure(figsize=(10, 4))
sns.countplot(x='Aggregate rating', data=df, palette='Set2')
plt.title('Count of Each Rating')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

*** Compare Average Ratings by Cuisine or City  ***

*   Average Ratings by Cuisine *

```
cuisine_ratings = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=cuisine_ratings.values, y=cuisine_ratings.index, palette='coolwarm')
plt.title('Top 10 Cuisines by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')
plt.show()
```

***   Average Ratings by City:***

```
city_ratings = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=city_ratings.values, y=city_ratings.index, palette='viridis')
plt.title('Top 10 Cities by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('City')
plt.show()
```

# Visualize Relationships Between Features and Rating

*   Online Delivery vs Rating*

```
plt.figure(figsize=(6, 4))
sns.boxplot(x='Has Online delivery', y='Aggregate rating', data=df, palette='pastel')
plt.title('Rating by Online Delivery Availability')
plt.show()
```

***  Table Booking vs Rating ***

```
plt.figure(figsize=(6, 4))
sns.boxplot(x='Has Table booking', y='Aggregate rating', data=df, palette='Set3')
plt.title('Rating by Table Booking Availability')
plt.show()
```

*** Price Range vs Rating  ***

```
plt.figure(figsize=(8, 5))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette='muted')
plt.title('Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Rating')
plt.show()
```

# Bonus: Heatmap of Correlations (if numerical columns exist)

```
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
